```
training_data = pd.concat([treatment_data_blank_instance,session_parameter_instance, reveal_data_blank_instance],axis=1)
```

```
training_data = training_data.drop(training_data.index[0])
```

```

for i in range(len(treatment_data)):
  treatment_instance = treatment_data.iloc[[i]]
  treatment_date_numeric_array = treatment_instance.iloc[[0],[5]]
  treatment_date_numeric_value = treatment_date_numeric_array.values[0][0]

  for j in range(len(session_parameters)):
    session_parameter_instance_read = session_parameters.iloc[[j]]
    #If we find a matching date for the session parmeters update it.
    # we shall that session parameters for the treatments thereafter
    session_date_numeric_array = session_parameter_instance_read.iloc[[0],[4]]
    session_date_numeric_value = session_date_numeric_array.values[0][0]
    if treatment_date_numeric_value == session_date_numeric_value:
      session_parameters_instance = session_parameter_instance_read.copy()
      break

  for k_index in range(len(reveal_data)):
    reveal_data_instance_read = reveal_data.iloc[[k_index]]
    reveal_date_numeric_array = reveal_data_instance_read.iloc[[0],[4]]
    reveal_date_numeric_value = reveal_date_numeric_array.values[0][0]
    if treatment_date_numeric_value == reveal_date_numeric_value:
      reveal_data_instance = reveal_data_instance_read.copy()
      break
    # Did not find a matching 
    if k_index == len(reveal_data) - 1:
      reveal_data_instance = reveal_data_blank_instance.copy()


  treatment_instance.reset_index(drop=True, inplace=True)
  session_parameters_instance.reset_index(drop=True, inplace=True)
  reveal_data_instance.reset_index(drop=True, inplace=True)
  training_instance = pd.concat([treatment_instance,session_parameters_instance, reveal_data_instance],axis=1)
  training_data = pd.concat([training_data,training_instance],axis=0)
  #training_data = training_data.append(training_instance)

print(training_data.shape)



```

```
writer = pd.ExcelWriter('M100100001.xlsx')
training_data.to_excel(writer)
writer.save()
```
